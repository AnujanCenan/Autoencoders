## Clean All Data
Since we can read in the PTB_XL data, we can use established preprocessing techniques to remove most common types of noise in an ECG

Apply this cleaning process to the data

```python
import sys
from pathlib import Path
import numpy as np

notebook_dir = Path().resolve()
data_reader_path = notebook_dir.parent/"data_reader"
plotter_path = notebook_dir.parent/"plotter"

pre_processor_path = notebook_dir.parent/"pre_processing"

if str(data_reader_path) not in sys.path:
    sys.path.append(str(data_reader_path))

if str(pre_processor_path) not in sys.path:
    sys.path.append(str(pre_processor_path))

if str(plotter_path) not in sys.path:
    sys.path.append(str(plotter_path))

from data_reader import PTB_XL_Reader
from pre_processor import Pre_Processor

data_reader = PTB_XL_Reader()
print("Reading in all data into RAM...")
num_records = data_reader.get_num_records()
X, fs = data_reader.get_all_raw_voltages()


pre_processor = Pre_Processor()

print("Cleaning training dataset...")
for i in range(num_records):
    X[i] = pre_processor.clean(X[i], fs)

np.save('cleaned_PTBXL_data.npy', X)
```
